```python
import plotly.express as px
import matplotlib.pyplot as plt
```

```python
fg = px.bar(df,x='Region',y=' Estimated Unemployment Rate (%)',color='Region',title='Unemployment rate',animation_frame=' Date',template='plotly')
fg.update_layout(xaxis={'categoryorder':'total descending'})
# Show the bar chart
fg.show()
```

```python
fg = px.box(df,x='Region',y=' Estimated Unemployment Rate (%)',color='Region',title='Unemployment rate',animation_frame=' Date',template='plotly')
fg.update_layout(xaxis={'categoryorder':'total descending'})
# Show the box chart
fg.show()
```

```python
fg = px.scatter(df,x='Region',y=' Estimated Unemployment Rate (%)',color='Region',title='Unemployment rate',animation_frame=' Date',template='plotly')
fg.update_layout(xaxis={'categoryorder':'total descending'})
# Show the scatter chart
fg.show()
```

```python
fg = px.histogram(df,x='Region',y=' Estimated Unemployment Rate (%)',color='Region',title='Unemployment rate',animation_frame=' Date',template='plotly')
fg.update_layout(xaxis={'categoryorder':'total descending'})
# Show the histogram chart
fg.show()
```

```python
fg = px.line(df,x='Region',y=' Estimated Unemployment Rate (%)',color='Region',title='Unemployment rate',animation_frame=' Date',template='plotly')
fg.update_layout(xaxis={'categoryorder':'total descending'})
# Show the line chart
fg.show()
```

```python
fig = px.pie(df, values=' Estimated Unemployment Rate (%)', names='Region', title='Unemployment rate by Region', template='plotly')
# Show the pie chart
fig.show()
```

```python
data = {
    'Category': ['A', 'B', 'C', 'D'],
    'X_Value': [10, 15, 8, 20],
    'Y_Value': [25, 18, 12, 30],
    'Bubble_Size': [100, 200, 300, 150]
}
fig = px.scatter(data_frame=data, x='X_Value', y='Y_Value', size='Bubble_Size', text='Category',
                 title='Bubble Chart', size_max=50)

# Show the bubble chart
fig.show()
```
